```python
import pathlib
import textwrap

import google.generativeai as genai

from IPython.display import display
from IPython.display import Markdown
from IPython.display import Audio, display

from gtts import gTTS

def to_markdown(text):
  text = text.replace('â€¢', '  *')
  return Markdown(textwrap.indent(text, '> ', predicate=lambda _: True))
```

```python
from google.colab import userdata
GOOGLE_API_KEY=userdata.get('Google_api_key')
genai.configure(api_key=GOOGLE_API_KEY)
model = genai.GenerativeModel('gemini-pro')
```

```python
# Step 1
def ask_question(name):
  ques = input(f"Hey {name}, what do you want? ")
  return ques
```

```python
ask_question("Deepesh")
```

```python
def classify_question(ques):
  goahead_with_web_search = False
  device_lst = ['alarm','reminder','message','call']
  personal_lst = ['who are you?', 'who created you']

  device = False
  for i in device_lst:
    if i in ques.lower():
      device = True

  if device:
    print('This question is related to Device things, which is not supported in our google assistant ')

  personal_question = False
  for i in personal_lst:
    if i in ques.lower():
      personal_question = True

  if personal_question:
    print("Hey am a personal assisstant created by Deepesh Ahuja")

  if device:
    goahead_with_web_search = False
  elif personal_question:
    goahead_with_web_search = False
  else:
    goahead_with_web_search = True

  return goahead_with_web_search


```

```python
classify_question("who are you?")
```

```python
# Start searching with google Gemini
```

```python
def ask_gemini(ques):
  modified_prompt = "hey give me answer to this question " + ques + "in maximum of 40 words"
  response = model.generate_content(modified_prompt)
  to_markdown(response.text)
  return response.text
```

```python
def speak(answer):
  tts = gTTS(answer)
  tts.save('audio.mp3')
  display(Audio("audio.mp3", autoplay=True))
```

```python
have_any_other_ques = 'y'
name=''

while have_any_other_ques == 'y':
  if name == '':
    name = input("Hey what is your name? - ")

  q = ask_question(name)

  go_ahead = classify_question(q)
  answer=''

  if go_ahead == True:
    answer = ask_gemini(q)
    speak(answer)
    print(answer)


  have_any_other_ques = input("Do you have any questions? ")
```
